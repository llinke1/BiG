```python
import numpy as np
from BiG import plotter as pl
import matplotlib.pyplot as plt
```

# Examples for plotting bispectra with BiG

## Loading data

```python
data=np.loadtxt("testRun_output_effectiveTriangles_all_test_grid_512_1.dat")

# Using effective triangle centers
ks=data[:,3:6]
bispec=data[:,8]
ks_lower=data[:,0:3]-0.05
ks_upper=data[:,0:3]+0.05
```

```python
Plots=pl.plotter()
Plots.addBispectrum(ks=ks, kbinedges_low=ks_lower, kbinedges_high=ks_upper, bispec=bispec, name='Bispectrum')
```

```python
np.min(np.log(bispec))
```

```python
np.max((bispec))
```

## Plotting bispectra as 3D scatter plot

```python
Plots.plotBispectrum3D(vmin=0, vmax=1e4)
```
