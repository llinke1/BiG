```python
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
import numpy as np

from BiG import plotter as pl
```

```python
Plots=pl.plotter()
```

```python
data=np.loadtxt("bahamas_theone_L512_Nmesh512_bcm.dat")
data2=np.loadtxt("bahamas_theone_L512_Nmesh512_gro.dat")
```

```python
k_data=data[:,:3]
```

```python
kbins=np.linspace(0.01, 1.91, 10)
kbins_lower=kbins[:-1]
kbins_upper=kbins[1:]
kbins_mid=0.5*(kbins_lower+kbins_upper)

ks_low=[]
ks_high=[]

for i in range(9):
    k1_low=kbins_lower[i]
    k1_high=kbins_upper[i]
    for j in range(i,9):
        k2_low=kbins_lower[j]
        k2_high=kbins_upper[j]
        for k in range(j,9):
            k3_low=kbins_lower[k]
            k3_high=kbins_upper[k]
            if kbins_mid[k]<=kbins_mid[i]+kbins_mid[j]:
                ks_low.append([k1_low, k2_low, k3_low])
                ks_high.append([k1_high, k2_high, k3_high])

ks_low=np.array(ks_low)
ks_high=np.array(ks_high)
```

```python
Plots.addBispectrum(ks=data[:,3:6], kbinedges_low=ks_low, kbinedges_high=ks_high, bispec=data[:,8], name='Gravity Only')
```

```python
Plots.addBispectrum(ks=data[:,3:6], kbinedges_low=ks_low, kbinedges_high=ks_high, bispec=data2[:,8], name='BCM')
```

```python
Plots.plotBispectrum3D(names=['Gravity Only'], vmin=np.log(1e5), vmax=np.log(5e7))
```

```python
Plots.plotBispectrum3D(names=['Gravity Only', 'BCM'], vmin=np.log(1e5), vmax=np.log(5e7))
```
